```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_fractions = np.array([0.25, 0.5, 0.75, 1])
```

```python
def sanitize_data(val_mAP, fractions):
    results = []
    for i in range(len(val_mAP)):
        if val_mAP[i] != -100:
            results.append([fractions[i], val_mAP[i]])

    return np.array(results)

def plot_two_series(series_a, series_b, series_a_label, series_b_label, title):
    plt.figure()
    plt.scatter(series_a[:,0], series_a[:,1], label=series_a_label, color='blue')
    plt.plot(series_a[:,0], series_a[:,1], color='blue')
    
    plt.scatter(series_b[:,0], series_b[:,1], label=series_b_label, color='red')
    plt.plot(series_b[:,0], series_b[:,1], color='red')
    plt.legend()

    plt.xlabel('Fraction of Cityscapes')
    plt.ylabel('Cityscapes Mask mAP')
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
mobilenetv2_imagenet_ssj_frac = np.array([0.221,0.2152,0.2231,0.2241])*100
mobilenetv2_imagenet_ssj_cp_frac = np.array([0.2317, 0.231, 0.2344, 0.2384])*100

mobilenetv2_imagenet_lsj_frac = np.array([0.2396, 0.2339, 0.2415, 0.2389])*100
mobilenetv2_imagenet_lsj_cp_frac = np.array([0.248, 0.2455, 0.2468, 0.2487])*100
```

```python
mobilenetv2_imagenet_ssj_frac_results = sanitize_data(mobilenetv2_imagenet_ssj_frac, dataset_fractions)
print(mobilenetv2_imagenet_ssj_frac_results)

mobilenetv2_imagenet_ssj_cp_frac_results = sanitize_data(mobilenetv2_imagenet_ssj_cp_frac, dataset_fractions)
print(mobilenetv2_imagenet_ssj_cp_frac_results)

mobilenetv2_imagenet_lsj_frac_results = sanitize_data(mobilenetv2_imagenet_lsj_frac, dataset_fractions)
print(mobilenetv2_imagenet_lsj_frac_results)

mobilenetv2_imagenet_lsj_cp_frac_results = sanitize_data(mobilenetv2_imagenet_lsj_cp_frac, dataset_fractions)
print(mobilenetv2_imagenet_lsj_cp_frac_results)
```

```python
plot_two_series(mobilenetv2_imagenet_ssj_frac_results, mobilenetv2_imagenet_ssj_cp_frac_results, 
                'SSJ', 'SSJ + Copy-Paste', title='MobileNetv2 FPN Backbone Pretrained on ImageNet')
```

```python
plot_two_series(mobilenetv2_imagenet_lsj_frac_results, mobilenetv2_imagenet_lsj_cp_frac_results, 
                'LSJ', 'LSJ + Copy-Paste', title='MobileNetv2 FPN Backbone Pretrained on ImageNet')
```
